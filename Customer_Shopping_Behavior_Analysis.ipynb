```python
import pandas as pd 
```

```python
df = pd.read_csv('ecommerce_sales_with_age.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df["Month"] = df["order_date"].dt.month_name()
```

```python
order_date_index = df.columns.get_loc("order_date")
cols = list(df.columns)
cols.insert(order_date_index + 1, cols.pop(cols.index("Month")))
df = df[cols]
```

```python
df.columns
```

```python
df["rating"] = df.groupby("category")["rating"].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df
```

```python
def age_group(age):
    if age < 18:
        return "Teenager"
    elif age <= 35:
        return "Young Adult"
    elif age <= 55:
        return "Adult"
    else:
        return "Senior"
```

```python
df["Age Group"] = df["Age"].apply(age_group)
```

```python
df.columns
```

```python
# Find index of Age column
age_index = df.columns.get_loc("Age")

# Reorder columns to place Age Group after Age
cols = list(df.columns)
cols.insert(age_index + 1, cols.pop(cols.index("Age Group")))
df = df[cols]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = pd.DataFrame(df)
df.to_csv('new_one.csv', index=False)
```
